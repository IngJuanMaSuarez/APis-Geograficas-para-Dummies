```python
import requests
import folium
```

```python
api_key = "0772b676ef44deb64c03aac472b12b48"
lat = 4.74
lon = -74.50

url = f"https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={api_key}"
```

```python
respuesta = requests.get(url)
data_clima = respuesta.json()
```

```python
mapa_colombia = folium.Map(location=(4.74, -74.50), zoom_start=5)
```

```python
clima = data_clima["weather"][0]["description"]
temperatura = data_clima["main"]["temp"] - 273.15
texto_popup = f"Clima: {clima} <br>Temperatura: {temperatura:.2f} °C"

folium.Marker(
    location = [lat, lon],
    popup = texto_popup,
    icon = folium.Icon(icon="cloud"),
).add_to(mapa_colombia)
```
